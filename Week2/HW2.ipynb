# Part 3: Evaluation Based on 4-Levels of Design by ChatGPT with Prompt Engineering Techniques

### 1. **Domain Situation**
   - **Context**: This visualization is likely aimed at an audience interested in global energy trends, especially those focused on solar energy production. This could include policy makers, energy researchers, and industry stakeholders who want to understand global manufacturing patterns and country-specific contributions in solar photovoltaic (PV) production.
   - **Audience Needs**: The audience needs to comprehend how global PV production is distributed across different countries and regions, observe production trends over time, and identify dominant players. They may also be interested in evaluating the concentration of PV production, especially noting how China's dominance has evolved.

### 2. **Data/Task Abstraction**
   - **Data Representation**: The chart uses a stacked bar graph format to depict annual solar photovoltaic cell production (in gigawatts) by country or region from 2010 to 2021. Each segment in the stacked bars represents a country's contribution to total global production for that year.
   - **Tasks**: The primary tasks for the audience are:
     - Identify the major contributing countries to PV production each year.
     - Observe trends over time, especially shifts in production dominance.
     - Compare the production output of China versus other regions.

### 3. **Visual Encoding/Interaction Idiom**
   - **Visual Encoding**: Each color represents a different country or region, and the bar height indicates total global PV production for each year. The size of each segment within a bar indicates the contribution of each country.
   - **Strengths**:
     - **Stacked Bar Chart**: Effective for comparing total production over time and seeing Chinas increasing dominance.
     - **Color Coding**: The distinct colors help in identifying the different countries.
   - **Limitations**:
     - **Comparing Smaller Sections**: The segments representing the United States, Europe, and other regions are quite small and close in color. This makes it challenging to differentiate or accurately compare their contributions, especially in recent years where China's segment is disproportionately large.
     - **Limited Trend Visibility for Non-Dominant Regions**: Due to Chinas overwhelming contribution, its hard to discern production trends for other regions, as these are represented by very small, hard-to-compare segments.

### 4. **Qualitative and Quantitative Analysis**
   - **Qualitative Feedback**: 
     - The visualization clearly highlights Chinas dominant position in solar PV production, but the nuances in contributions from other countries or regions are largely lost, making it challenging to analyze trends for these smaller contributors.
     - The legend is clear, and the visualization is visually appealing. However, the intended message of global production patterns is skewed due to the overwhelming focus on China.
   - **Quantitative Usability Metrics**:
     - **Readability**: High for understanding Chinas dominance but low for identifying contributions of other countries, especially the United States, Europe, and Taiwan.
     - **Trend Detection**: Chinas trend is easy to detect, but trends of smaller segments are not easily observable without detailed labels or a more precise means of comparison.

---

### **Two Key Areas for Improvement**
   1. **Increase Comparability for Smaller Countries/Regions**:
      - Consider a separate visualization or secondary chart (such as a line chart) for non-China regions. This would allow for a clearer view of trends and comparisons among countries with smaller contributions.
   
   2. **Enhance Detail on Smaller Contributions**:
      - Adding labels to each segment with percentage contributions or specific production values could help make the data more interpretable.
      - Alternatively, a split-axis design or an inset chart focused on non-China contributions could improve readability without diminishing the prominence of Chinas trend.


### Revised design

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Simulated data
simulated_data = {
    "Year": np.arange(2010, 2022),
    "China (GW)": [10, 20, 30, 45, 60, 80, 100, 120, 130, 140, 160, 170],
    "Taiwan (GW)": [5, 7, 8, 10, 10, 12, 12, 13, 13, 13, 12, 11],
    "Malaysia (GW)": [4, 5, 6, 6, 8, 10, 10, 10, 10, 10, 10, 10],
    "Japan (GW)": [5, 5, 5, 6, 8, 8, 8, 7, 6, 6, 6, 5],
    "Europe (GW)": [6, 7, 8, 9, 10, 11, 10, 9, 8, 7, 6, 5],
    "United States (GW)": [8, 9, 10, 11, 12, 12, 12, 11, 10, 9, 8, 7],
    "ROW (GW)": [2, 3, 5, 7, 10, 15, 20, 25, 30, 35, 40, 50]
}
colors = ["#D62728", "#8C564B", "#FF7F0E", "#9467BD", "#2CA02C", "#1F77B4", "#7F7F7F"]

# Create DataFrame
data = pd.DataFrame(simulated_data)

# Stacked Bar Chart
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 12), gridspec_kw={'height_ratios': [2, 1]})

# Main Stacked Bar Chart (Top Chart)
bottom = np.zeros(len(data["Year"]))
for column, color in zip(data.columns[1:], colors):
    bars = ax1.bar(data["Year"], data[column], bottom=bottom, label=column, color=color)
    # Add labels to each bar segment
    for bar in bars:
        height = bar.get_height()
        if height > 0:  # Only add label if height is positive
            ax1.text(
                bar.get_x() + bar.get_width() / 2,
                bar.get_y() + height / 2,
                f'{int(height)}',
                ha='center', va='center', fontsize=8, color='white'
            )
    bottom += data[column]  # Update bottom to stack next region

# Labels and Title for Main Chart
ax1.set_xlabel("Year")
ax1.set_ylabel("Annual solar photovoltaic cell production (GW)")
ax1.set_title("Annual Solar Photovoltaic Cell Production by Region (2010-2021)")
ax1.legend(title="Origin", loc="upper left", bbox_to_anchor=(1, 1))
ax1.set_xticks(data["Year"])

# Secondary Line Chart for Non-China Regions (Bottom Chart)
for column, color in zip(data.columns[2:], colors[1:]):  # Skip "China (GW)"
    ax2.plot(data["Year"], data[column], label=column, color=color, marker='o')

# Labels and Title for Secondary Chart
ax2.set_xlabel("Year")
ax2.set_ylabel("Production (GW)")
ax2.set_title("Annual Solar Photovoltaic Cell Production (Excluding China)")
ax2.legend(title="Origin", loc="upper left", bbox_to_anchor=(1, 1))
ax2.set_xticks(data["Year"])

plt.tight_layout()
plt.show()
```

#### An interesting alternation
- Use a Split-Axis Design
    - A split-axis or broken y-axis can be used to magnify the lower part of the chart. This way, the contributions from countries with lower production levels (like the U.S. and Europe) are visually expanded, making it easier to distinguish their contributions.

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Simulated data
simulated_data = {
    "Year": np.arange(2010, 2022),
    "China (GW)": [10, 20, 30, 45, 60, 80, 100, 120, 130, 140, 160, 170],
    "Taiwan (GW)": [5, 7, 8, 10, 10, 12, 12, 13, 13, 13, 12, 11],
    "Malaysia (GW)": [4, 5, 6, 6, 8, 10, 10, 10, 10, 10, 10, 10],
    "Japan (GW)": [5, 5, 5, 6, 8, 8, 8, 7, 6, 6, 6, 5],
    "Europe (GW)": [6, 7, 8, 9, 10, 11, 10, 9, 8, 7, 6, 5],
    "United States (GW)": [8, 9, 10, 11, 12, 12, 12, 11, 10, 9, 8, 7],
    "ROW (GW)": [2, 3, 5, 7, 10, 15, 20, 25, 30, 35, 40, 50]
}
colors = ["#D62728", "#8C564B", "#FF7F0E", "#9467BD", "#2CA02C", "#1F77B4", "#7F7F7F"]

# Create DataFrame
data = pd.DataFrame(simulated_data)

# Split-Axis Stacked Bar Chart
fig, (ax1, ax2) = plt.subplots(2, 1, sharex=True, figsize=(10, 8), gridspec_kw={'height_ratios': [2, 1]})

# Upper Chart for High Range (for larger values, like China's contribution)
bottom = np.zeros(len(data["Year"]))
for column, color in zip(data.columns[1:], colors):
    ax1.bar(data["Year"], data[column], bottom=bottom, label=column, color=color)
    bottom += data[column]  # Update bottom to stack next region

# Lower Chart for Low Range (magnifying the smaller values)
bottom = np.zeros(len(data["Year"]))
for column, color in zip(data.columns[1:], colors):
    ax2.bar(data["Year"], data[column], bottom=bottom, label=column, color=color)
    bottom += data[column]

# Set y-axis limits to create the split effect
ax1.set_ylim(100, 250)  # Upper limit for higher values
ax2.set_ylim(0, 50)     # Lower limit for smaller values

# Adjust aesthetics
ax1.spines['bottom'].set_visible(False)
ax2.spines['top'].set_visible(False)
ax1.xaxis.tick_top()
ax1.tick_params(labeltop=False)  # Don't put tick labels at the top
ax2.xaxis.tick_bottom()

# Add a diagonal line at the split for visual continuity
d = .015  # Diagonal line size
kwargs = dict(transform=ax1.transAxes, color='k', clip_on=False)
ax1.plot((-d, +d), (-d, +d), **kwargs)
ax1.plot((1 - d, 1 + d), (-d, +d), **kwargs)
kwargs.update(transform=ax2.transAxes)  # Switch to the lower axis
ax2.plot((-d, +d), (1 - d, 1 + d), **kwargs)
ax2.plot((1 - d, 1 + d), (1 - d, 1 + d), **kwargs)

# Labels and Title
ax2.set_xlabel("Year")
ax1.set_ylabel("Annual solar photovoltaic cell production (GW)")
ax2.set_ylabel("Annual solar photovoltaic cell production (GW)")
ax1.legend(title="Origin", loc="upper left", bbox_to_anchor=(1, 1))

plt.xticks(data["Year"])
plt.tight_layout()
plt.show()
```
